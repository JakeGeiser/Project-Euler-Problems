<p>$2520$ is the smallest number that can be divided by each of the numbers from $1$ to $10$ without any remainder.</p>
<p>What is the smallest positive number that is <strong class="tooltip">evenly divisible<span class="tooltiptext">divisible with no remainder</span></strong> by all of the numbers from $1$ to $20$?</p>

```python
def isDivisibleByRange(minVal,maxVal,productToCheck):
    if minVal<=maxVal:
        if productToCheck%minVal == 0:
            return isDivisibleByRange(minVal+1,maxVal,productToCheck)
        else: return True
    else: return False
    
```

```python
minRange = 1
maxRange = 5

isNotEvenlyDivisible = True

product = 0

while isNotEvenlyDivisible:
    product += maxRange
    isNotEvenlyDivisible = isDivisibleByRange(minRange,maxRange,product)
    
```
