<p>What is the largest prime factor of the number $600851475143$?</p>

```python
#Find largest prime factor of number N
N = 600851475143
#N = 51
```

```python
largestPrime = 1
prime = 2

# check for the only even prime number as a factor
if N%prime == 0:
    largestPrime = prime
    N = N/prime

# set prime to 3 so we can iterate on odd numbers
prime = 3

while N>1:
    if N%prime == 0:
        largestPrime = prime
        N = N/prime
        continue
    else:
        prime += 2
        
    
```
